# Extract Overall Summary Content from Model Outputs

This notebook extracts only the **overall summary** content from TextRank, LexRank, and DistilBART output files, removing titles and statistics sections.

The cleaned overall summaries are saved to the `OP_SUMMARIES` folder.


```python
import os
import re
from pathlib import Path

# Configuration
base_path = Path(".")
textrank_results = base_path / "textrank_results"
lexrank_results = base_path / "lexrank_results"
distilbart_results = base_path / "distilbart_results"
output_folder = base_path / "OP_SUMMARIES"

# Create output folder
output_folder.mkdir(exist_ok=True)
```

```python
def extract_summary_content(file_path):
    """
    Extract only the summary content from a summary file.
    Removes title, separators, and statistics sections.
    """
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    summary_lines = []
    found_first_separator = False
    found_content = False
    
    for line in lines:
        line_stripped = line.strip()
        
        # Skip title line
        if 'Summary' in line_stripped and ('TextRank' in line_stripped or 'LexRank' in line_stripped or 'Abstractive' in line_stripped or 'DistilBART' in line_stripped):
            continue
        
        # Skip title line for yearly summaries
        if line_stripped.startswith('Mercedes F1 Infringement Summary'):
            continue
        
        # Detect first separator
        if line_stripped.startswith('=') and not found_first_separator:
            found_first_separator = True
            continue
        
        # Stop at second separator or STATISTICS section
        if found_first_separator and (line_stripped.startswith('=') or line_stripped.startswith('STATISTICS')):
            break
        
        # Collect content after first separator
        if found_first_separator and line_stripped:
            found_content = True
            # Remove year prefixes like "2020 Summary:" from overall summaries
            cleaned_line = re.sub(r'^\d{4}\s+Summary:\s*', '', line_stripped)
            if cleaned_line:
                summary_lines.append(cleaned_line)
    
    return ' '.join(summary_lines).strip()
```

```python
# Process TextRank overall summary
print("Processing TextRank overall summary...")
input_file = textrank_results / "textrank_overall_summary.txt"
if input_file.exists():
    summary_content = extract_summary_content(input_file)
    output_file = output_folder / "textrank_overall_summary.txt"
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(summary_content)
    print(f"  ✓ Processed overall summary ({len(summary_content)} characters)")
else:
    print(f"  ✗ File not found: {input_file}")

print("\n" + "="*60 + "\n")
```

```python
# Process LexRank overall summary
print("Processing LexRank overall summary...")
input_file = lexrank_results / "lexrank_overall_summary.txt"
if input_file.exists():
    summary_content = extract_summary_content(input_file)
    output_file = output_folder / "lexrank_overall_summary.txt"
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(summary_content)
    print(f"  ✓ Processed overall summary ({len(summary_content)} characters)")
else:
    print(f"  ✗ File not found: {input_file}")

print("\n" + "="*60 + "\n")
```

```python
# Process DistilBART overall summary
print("Processing DistilBART overall summary...")
input_file = distilbart_results / "distilbart_overall_summary.txt"
if input_file.exists():
    summary_content = extract_summary_content(input_file)
    output_file = output_folder / "distilbart_overall_summary.txt"
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(summary_content)
    print(f"  ✓ Processed overall summary ({len(summary_content)} characters)")
else:
    print(f"  ✗ File not found: {input_file}")

print("\n" + "="*60 + "\n")
```

```python
# Summary
print("Summary of extracted overall summaries:")
print("="*60)
output_files = sorted(output_folder.glob("*_overall_summary.txt"))
print(f"\nTotal overall summary files created: {len(output_files)}")
print("\nOverall summaries in OP_SUMMARIES folder:")
for file in output_files:
    file_size = file.stat().st_size
    print(f"  - {file.name} ({file_size:,} bytes)")
```
